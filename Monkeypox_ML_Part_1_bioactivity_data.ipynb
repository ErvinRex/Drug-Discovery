```python
# import relevant libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#Search for target protein
target = new_client.target
target_query = target.search('monkeypox')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#Select and retrieve bioactivity data for monkeypox in humans
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
res
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
#note standard value = concentration of drug required to account for 50% inhibition
```

```python
df.standard_type.unique()
```

```python
df.to_csv('bioactivity_data.csv', index=False)
```

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
#Labeling compounds as either being active, inactive or intermediate
#Compounds having values of less than 1000 nM will be considered to be **active** while those greater than 10,000 nM will be considered to be **inactive**. 
#As for those values in between 1,000 and 10,000 nM will be referred to as **intermediate**. 
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
bioactivity_class
```

```python
#Iterate to get molecule_chemble_id, canonical_smiles & standard_value
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
df3
```

```python
#attach bioactivity_class
bioactivity_class = pd.Series(bioactivity_class, name = 'bioactivity_class')
df4 = pd.concat([df3, pd.Series(bioactivity_class)], axis=1)
df4
```

```python
df4.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
